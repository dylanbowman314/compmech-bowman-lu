```python
import pickle
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, Normalize
import numpy as np
```

```python
with open("r2.pkl", "rb") as f:
    r2_values = pickle.load(f)
```

```python
r2_015_06 = r2_values[list(r2_values.keys())[0]]
r2_005_085 = r2_values[list(r2_values.keys())[1]]
```

```python
unique_xa = set([(x, a) for (x, a) in list(r2_015_06.keys())])
```

```python
x_points = np.zeros([len(unique_xa)])
a_points = np.zeros([len(unique_xa)])
r2_for_xa = np.zeros([len(unique_xa)])
for i, (x, a) in enumerate(r2_015_06.keys()):
    x_points[i] = x
    a_points[i] = a
    r2_for_xa[i] = r2_015_06[(x, a)].item()
```

```python
x_points = np.zeros([len(unique_xa)])
a_points = np.zeros([len(unique_xa)])
r2_for_xa_085 = np.zeros([len(unique_xa)])
for i, (x, a) in enumerate(r2_005_085.keys()):
    x_points[i] = x
    a_points[i] = a
    r2_for_xa_085[i] = r2_005_085[(x, a)].item()
```

```python
fig, ax = plt.subplots()
cmap = plt.cm.viridis
scatter = ax.scatter(x_points, a_points, c=r2_for_xa, cmap=cmap, s=5, norm=Normalize(vmin=0.98, vmax=1.0))

cbar = fig.colorbar(scatter, ax=ax)
cbar.ax.set_ylabel('r^2', rotation=-90, va='bottom')
ax.set_xlabel('x')
ax.set_ylabel('a')
ax.set_title('r^2 for linear probe on transformer trained on x = 0.15, a = 0.6, vs. true probabilities')

plt.show()
```
